# Bayesian network where a detector can fire multiple times

```python
import matplotlib.pyplot as plt
```

```python
def p_indicator(p_detector, N):
    """Probability that the indicator is true."""
    
    # Preconditions
    assert isinstance(p_detector, float)
    assert 0 <= p_detector <= 1, f"Invalid probability: {p_detector}"
    assert isinstance(N, int)
    assert N >= 0
    
    return 1 - (1 - p_detector)**N
```

```python
n = list(range(0, 10))
p = [p_indicator(0.7, ni) for ni in n]
```

```python
plt.plot(n, p)
plt.xlabel("Number of occurrences")
plt.ylabel("Probability of indicator")
plt.show()
```
