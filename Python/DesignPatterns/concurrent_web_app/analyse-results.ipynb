# Analyse Experiment 2 results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('experiment_2_results.pkl')
df.head()
```

```python
df['n_unique_pids'].plot(kind='hist')
plt.xlabel('Number of unique processes')
plt.savefig('experiment_2_hist.png')
```

```python
df.plot.scatter(x='n_unique_pids', y='time_taken', s=2)
plt.xlabel('Number of unique processes')
plt.ylabel('Time taken (seconds)')
plt.savefig('experiment_2_results.png')
```
